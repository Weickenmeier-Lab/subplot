```python
def data_analysis(exp_num, file_name):
    data_file = pd.read_excel(file_name)
    experiment = data_file[data_file.Column2 == exp_num]
    disp = np.array(experiment.Column4)
    force = np.array(experiment.Column12)
    for m, n in enumerate(force):
        if n == np.amax(force):
            break
    disp_loading = disp[10:m]
    force_loading = force[10:m]
    for i,val in enumerate(disp_loading):
        slope_all=[]   
        if i<=10:
            continue
        for j in range(i-10, i):
            delx=disp_loading[j]-disp_loading[j-1]
            dely=force_loading[j]-force_loading[j-1]
            slope=dely/delx
            slope_all.append(slope)
        slope_mean=np.mean(slope_all)
        if slope_mean>0.02:
            break
    location=i-10
    disp_loading_new=disp_loading[location:]-disp_loading[location]
    force_loading_new=force_loading[location:]-force_loading[location]
    for o, p in enumerate(disp_loading_new):
        if p >= 50:
            break
        else:
            p = np.amax(disp_loading_new)
    disp_loading_origin=disp_loading_new[:o]
    force_loading_origin=force_loading_new[:o]
    plt.figure(1)
    plt.plot(disp_loading_origin, force_loading_origin, label=exp_num)
    plt.xlabel("Displacement (um)")
    plt.ylabel("Force (uN)")
    plt.title("3cc")
    plt.savefig("3cc_slope")
row = 36
for i in range(row):
    data_analysis(i, "3cc.xlsx")
```
